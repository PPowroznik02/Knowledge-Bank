```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Transforming DataFrames


#### Basics


```python
temperatures = pd.read_csv('data/temperatures.csv')
```

```python
# Returns first few rows of dataframe
print(temperatures.head())

# Returns shape of dataframe
print(temperatures.shape)

# Returns elementary statictics for values in colums
print(temperatures.describe())

# Returns basic information about non-null values and data types
print(temperatures.info())
```

```python
# Returns values in 2d numpy array
print(temperatures.values)

# Returns all column names in dataframe
print(temperatures.columns)

# Returns index of dataframe
print(temperatures.index)
```

#### Sorting and subsetting


```python
# Sorts rows first by temperature and next by date. Sorting by temperature is in acsending order, by date in descending.
print(temperatures.sort_values(['avg_temp_c', 'date'], ascending=[True, False]))
```

```python
# Selects specific columns from dataframe
print(temperatures[['avg_temp_c', 'date']])
```

```python
# Applys logical condiction on rows for values in specific column
print(temperatures['avg_temp_c'] > 0)
```

```python
# Select rows where condition is true
print(temperatures[temperatures['avg_temp_c'] > 0])
```

```python
countries = ['China', 'Poland']

# Selects columns in list
print(temperatures[temperatures.country.isin(countries)])
```

```python
# Adda new column with temperature in Fahrenheits
temperatures['avg_temp_f'] = (temperatures['avg_temp_c'] * 9/5) +32

print(temperatures.head())
```

### Aggregating DataFrames


```python
# Summary statistics
print('Mean: ', temperatures['avg_temp_c'].mean())
print('Median: ', temperatures['avg_temp_c'].median())
print('Mode: ', temperatures['country'].mode())
print('Min: ', temperatures['avg_temp_c'].min())
print('Max: ', temperatures['avg_temp_c'].max())
print('Var: ', temperatures['avg_temp_c'].var())
print('Std: ', temperatures['avg_temp_c'].std())
print('Sum: ', temperatures['avg_temp_c'].sum())
print('Quantile: ', temperatures['avg_temp_c'].quantile())
```

```python
# Aggregate method
def pct30(column):
    return column.quantile(0.3)

# 30 percentile of columns
print(temperatures[['avg_temp_c', 'avg_temp_f']].agg(pct30))
```

```python
# Cumulative statistics
temperatures['avg_temp_c'].cumsum()
temperatures['avg_temp_c'].cummax()
temperatures['avg_temp_c'].cummin()
temperatures['avg_temp_c'].cumprod()
```

```python
# Drops with duplicate city name values
temperatures.drop_duplicates(subset='city')
```

```python
# Counting categorical variable
print(temperatures['country'].value_counts())
print(temperatures['country'].value_counts(normalize=True))
```

```python
# Summaries by group
print(temperatures.groupby('country')['avg_temp_c'].min())
```

```python
# Pivot table, values: column to summarize, index: column to groupby
temperatures.pivot_table(values='avg_temp_c', index='country', aggfunc=[np.mean, np.median])
```

### Slicing and Indexing DataFrames


```python
# Sets index based on city colum
temperatures = temperatures.set_index('city')

# Selects rows where Xian and Abidjan is in city index
print(temperatures.loc[['Xian', 'Abidjan']])
```

```python
# Sorts rows by index
temperatures.sort_index(level='city', ascending=True)
```

```python
# Resets index and drop its content
temperatures.reset_index(drop=True)

# Resets index in dataframe
temperatures = temperatures.reset_index()
```

```python
temperatures = temperatures.set_index('date').sort_index()

# Slicing column names
temperatures.loc[:, 'country':'avg_temp_c']

# Slicing rows by index
temperatures.loc['2009-02-01':'2011-02-01', :]
```

```python
# Slicing to get the first 5 rows
print(temperatures.iloc[:5])

# Slicing to get columns 3 to 4
print(temperatures.iloc[:, 2:4])

# Slicing in both directions at once
print(temperatures.iloc[:5, 2:4])

temperatures = temperatures.reset_index()
```

```python
# Pivoting tables
temperatures = temperatures = pd.read_csv(
    'data/temperatures.csv',
    parse_dates=['date'] # Convert to date
)
temperatures['year'] = temperatures['date'].dt.year

# Pivot avg_temp_c by country and city vs year
print(temperatures.pivot_table(
            'avg_temp_c', 
            index=['country', 'city'],
            columns='year')
)

```
